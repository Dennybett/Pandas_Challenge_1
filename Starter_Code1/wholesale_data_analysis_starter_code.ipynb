## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
import pandas as pd

df['subtotal'] = df['unit_price'] * df['qty']
# df[['subtotal']]
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
import pandas as pd
import numpy as np

# df['shipping price'] = np.where(df['total_weight'] > 50, df['total_weight'] * 7, df['total_weight'] * 10)
df['total_weight'] = df['qty'] * df['unit_weight']
def shipping_calculator(weight):
    if weight > 50:
        return weight * 7
    return weight * 10

df['shipping_price'] = df['total_weight'].apply(shipping_calculator)
df[['shipping_price', 'total_weight', 'qty', 'unit_weight', 'unit_price', 'subtotal']]

```

```python
print(df.columns)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
import pandas as pd

df['total_price'] = (df['subtotal'] + df['shipping_price']) * (1 + (9.25 / 100))
df[['total_price', 'subtotal', 'shipping_price']].head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
import pandas as pd

df['cost_of_each_line'] = (df['unit_cost'] * df['qty'] + df['shipping_price'])
print(df.columns) 
```

```python
# Create a column for the profit of each line using line cost and line price
import pandas as pd

df['profit_of_each_line'] = df['total_price'] - df['cost_of_each_line']
print(df.columns)
```

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

filtered_df = df[df['client_id'].isin(most_entries_clients)]
top_5_clients_expenditure = filtered_df.groupby('client_id')['total_price'].sum().reset_index()
top_5_clients_expenditure = top_5_clients_expenditure.merge(df[['first', 'last', 'client_id']],on='client_id', how='left').drop_duplicates()
top_5_clients_expenditure = top_5_clients_expenditure.sort_values(by='total_price', ascending=False)
top_5_clients_expenditure = top_5_clients_expenditure[['first', 'last', 'client_id', 'total_price']]
top_5_clients_expenditure
```

```python
# Create a summary DataFrame showing the totals for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
# total units purchased = qty * item_id

import pandas as pd
summary_dataframe = df[df['client_id'].isin(most_entries_clients)]
summary_dataframe = summary_dataframe.groupby('client_id').agg({'qty': 'sum',
    'shipping_price': 'sum',
    'cost_of_each_line': 'sum',
    'profit_of_each_line': 'sum'
}).reset_index()
summary_dataframe
```

```python
summary_dataframe.dtypes
```

```python
# Format the data and rename the columns to names suitable for presentation.
import pandas as pd

summary_dataframe.rename(columns={'client_id': 'client_ID', 'qty': 'quantity', 'shipping_price': '$total_shipping_price',
                                 'cost_of_each_line': '$total_revenue', 'profit_of_each_line': '$total_profit'},inplace=True)



# Define the money columns. 
summary_dataframe.dtypes
# Define a function that converts a dollar amount to millions.
def conversion_to_millions(dollars):
        if isinstance(dollars, (float)):
            return dollars / 1000000
        else:
            return None
        
    
        
# Apply the currency_format_millions function to only the money columns. 
money_columns = ['$total_shipping_price', '$total_revenue', '$total_profit']
summary_dataframe[money_columns] = summary_dataframe[money_columns].applymap(conversion_to_millions)
summary_dataframe

# Rename the columns to reflect the change in the money format. 
summary_dataframe.rename(columns={'$total_shipping_price': '$total_shipping_price(millions)',
                                 '$total_revenue': '$total_revenue(millions)', '$total_profit': '$total_profit(millions)'},inplace=True)
summary_dataframe

```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
sorted_by_millions = summary_dataframe.sort_values(by='$total_profit(millions)', ascending=False)
sorted_by_millions
```
